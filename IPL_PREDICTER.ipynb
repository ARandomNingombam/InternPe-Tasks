# DATA PREPROCESSING

```python
total_score=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score=total_score[total_score['inning'] == 1]
```

```python
matches.head()
```

```python
match_data=matches.merge(total_score[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_data
```

```python
match_data['team1']=match_data['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_data['team2']=match_data['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
match_data
```

```python
match_data['team1'].unique()
```

```python
teams = [
     'Chennai Super Kings',
     'Delhi Capitals',
     'Gujarat Titans',
     'Kolkata Knight Riders',
     'Lucknow Super Giants',
     'Mumbai Indians',
     'Punjab Kings',
     'Rajasthan Royals',
     'Royal Challengers Bengaluru',
     'Sunrisers Hyderabad'
]
```

```python
match_data=match_data[match_data['team1'].isin(teams)]
```

```python
match_data=match_data[match_data['team2'].isin(teams)]
```

```python
match_data
```

```python
match_data.columns
```

```python
match_data=match_data[['match_id','city','winner','total_runs']]
```

```python
delivery_data=match_data.merge(delivery,on='match_id')
```

```python
delivery_data
```

```python
delivery_data=delivery_data[delivery_data['inning']==2]
```

```python
delivery_data.columns
```

```python
delivery_data
```

```python
delivery_data.info()
```

```python
delivery_data['current_score']=delivery_data.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_data
```

```python
delivery_data['runs_left_to_score']=delivery_data['total_runs_x']-delivery_data['current_score']
```

```python
delivery_data['runs_left_to_score']
```

```python
delivery_data['balls_left'] = 120 - (delivery_data['over'] * 6 + delivery_data['ball'])
```

```python
delivery_data
```

```python
delivery_data['player_dismissed']=delivery_data['player_dismissed'].fillna('0')
delivery_data['player_dismissed']=delivery_data['player_dismissed'].apply(lambda x:x if x=='0' else '1')
delivery_data['player_dismissed']=delivery_data['player_dismissed'].astype('int')
```

```python
wickets=delivery_data.groupby('match_id')['player_dismissed'].cumsum().values
```

```python
delivery_data['wickets_left']=10-wickets
```

```python
delivery_data
```

```python
delivery_data.tail()
```

```python
delivery_data.columns
```

```python
delivery_data=delivery_data[delivery_data['balls_left']!=0]
```

```python
delivery_data['crr']=(delivery_data['current_score']*6)/(120-delivery_data['balls_left'])
```

```python
delivery_data['rrr']=(delivery_data['runs_left_to_score']*6)/(delivery_data['balls_left'])
```

```python
def result(row):
  return 1 if row['batting_team']==row['winner'] else 0
```

```python
delivery_data['result']=delivery_data.apply(result,axis=1)
```

```python
final_data=delivery_data[['batting_team','bowling_team','city','runs_left_to_score','balls_left','total_runs_x','crr','rrr','result']]
```

```python
final_data=final_data.sample(final_data.shape[0])
```

```python
final_data.dropna(inplace=True)
```

```python
final_data=final_data[final_data['balls_left']!=0]
```

# DATA SPLITTING AND FEATURE ENGINEERING

```python
from sklearn.model_selection import train_test_split
```

```python
x=final_data.iloc[:,:-1]
y=final_data.iloc[:,-1]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
xtrain
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough'
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline as p
```

```python
pipe=p(
    steps=[
        ('step1',trf),
        ('step2',LogisticRegression(solver='liblinear'))
    ]
)
```

```python
pipe.fit(xtrain,ytrain)
```

```python
xtrain.describe()
```

```python
ypred=pipe.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypred)
```

# WE HAVE ACHEIVED A SCORE OF 81% ACCURACY WITH LOGISTIC REGRESSION

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest_pipe=p(
    steps=[
        ('step1',trf),
        ('step2',RandomForestClassifier())
    ]
)
```

```python
forest_pipe.fit(xtrain,ytrain)
```

```python
forest_ypred=pipe.predict(xtest)
```

# A score of 81% accuracy was achieved here as well using RandomForest

```python
accuracy_score(ytest,forest_ypred)
```
